# Q2 make trackbar to control line position ,color and thickness in image 500*500 rgb

```python
cv2.namedWindow("image")
def nothing(x):
    pass
cv2.createTrackbar('x0','image',0,499,nothing)
cv2.createTrackbar('x1','image',0,499,nothing)
cv2.createTrackbar('y0','image',0,499,nothing)
cv2.createTrackbar('y1','image',0,499,nothing)
cv2.createTrackbar('thick','image',1,10,nothing)
cv2.createTrackbar("R","image",0,255,nothing)
cv2.createTrackbar("G","image",0,255,nothing)
cv2.createTrackbar("B","image",0,255,nothing)
while True:
    img=np.zeros((500,500,3),np.uint8)
    x0=cv2.getTrackbarPos('x0','image')
    x1=cv2.getTrackbarPos('x1','image')
    y0=cv2.getTrackbarPos('y0','image')
    y1=cv2.getTrackbarPos('y1','image')
    R=cv2.getTrackbarPos('R','image')
    G=cv2.getTrackbarPos('G','image')
    B=cv2.getTrackbarPos('B','image')
    thick_2=cv2.getTrackbarPos('thick','image')
    cv2.line(img,(x0,y0),(x1,y1),(R,G,B),thick_2)
    cv2.imshow('my_image',img[:,:,::-1])
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
```

# Q3 make trackbar to control position,radius and color of circle in image 500*500 rgb.

```python
cv2.namedWindow("image")
def nothing(x):
    pass
cv2.createTrackbar('xo','image',0,499,nothing)
cv2.createTrackbar('yo','image',0,499,nothing)
cv2.createTrackbar('R','image',0,255,nothing)
cv2.createTrackbar('G','image',0,255,nothing)
cv2.createTrackbar('B','image',0,255,nothing)
cv2.createTrackbar('radius','image',0,499,nothing)

while True:
    img_3=np.zeros((500,500,3),np.uint8)
    x=cv2.getTrackbarPos('xo','image')
    y=cv2.getTrackbarPos('yo','image')
    rad=cv2.getTrackbarPos('radius','image')
    r=cv2.getTrackbarPos('R','image')
    g=cv2.getTrackbarPos('G','image')
    b=cv2.getTrackbarPos('B','image')
    cv2.circle(img_3,(x,y),rad,(r,g,b),2)
    cv2.imshow('image',img_3[:,:,::-1])
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
```

# Q4 create image 800*800 rgb write a word “Hello”with red color and draw yellow rectangle on border of word

```python
img_4=np.zeros((800,800,3),np.uint8)
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img_4,"Hello",(100,250),font,4,(255,0,0),5)
cv2.rectangle(img_4,(100,255),(430,155),(0,0,255),2)
plt.imshow(img_4)
```
