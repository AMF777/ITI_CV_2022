# Q10- read “Task_in” image and then print image shape ,plot the image using  matplotlib and opencv then edit the right corner pixel to be red color then save the image as (.jpg,.npg) to your current folder name it “my_image.png” then read both image again. “Make your observation”

```python
# import packages 
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# reading image
img=cv2.imread("Task_in.jpg")#BGR
plt.figure(figsize=(10,10))
plt.imshow(img[:,:,::-1])#BGR => RGB
print(img.shape)# to know the size of the np array
# change the bottom right corner to red 
```

```python
# change the top right corner to red 
img[0][20]=[255,0,0]
print(img)
plt.figure(figsize=(10,10))
plt.imshow(img)
cv2.imwrite("my_image.png",img[:,:,::-1])#to be saved as RGB NOT BGR
```

```python
img2=cv2.imread("my_image.png")
img2=img2[:,:,::-1]
plt.figure(figsize=(10,10))
plt.imshow(img2)
```

# Observation :  The main difference between JPG and PNG is the compression algorithms that they use. JPG uses a lossy compression algorithm that discards some of the image information in order to reduce the size of the file. In comparison, PNG uses a lossless algorithm that keeps all the information. With PNG, the quality of the image will not change, but the size of the file will usually be larger. On the other hand, JPG images can be made very small, but the quality can degrade very quickly from a certain point.


# Q11- make 51*51*3 RGB image with Blue color then save it as”Blue.jpg”

```python
rgbArray = np.zeros((51,51,3), 'uint8')
plt.figure(figsize=(10,10))
plt.title("before")
plt.imshow(rgbArray) #RGB
rgbArray.shape
```

```python
rgbArray[:]=[0,0,255]
print(rgbArray)
plt.figure(figsize=(10,10))
plt.title("After")
plt.imshow(rgbArray)
```

```python
# save image 
cv2.imwrite("Blue.jpg",rgbArray[:,:,::-1])#to be saved as RGB NOT BGR
```

# Q12- Make trackbar for rgb colors with window “colors” 

```python

def nothing(x):
    print(x)
cv2.namedWindow("my_image")#
cv2.createTrackbar('R','my_image',0,255,nothing)
cv2.createTrackbar('G','my_image',0,255,nothing)
cv2.createTrackbar('B','my_image',0,255,nothing)
rgb=np.zeros((500,500,3),np.uint8)# to create 3D numpy array that would represent image
while True:
    r=cv2.getTrackbarPos('R','my_image')
    g=cv2.getTrackbarPos('G','my_image')
    b=cv2.getTrackbarPos('B','my_image')
    rgb[:]=[r,g,b]#to make the image adapt new values made using track bar
    cv2.imshow("my_image",rgb[:,:,::-1])# RGB
    print(rgb)
    k=cv2.waitKey(1000)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
```
